```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data (replace this with your actual dataset)
data = {
    'Category': ['Normal', 'Generic', 'Exploits', 'DoS', 'Reconnaissance'],
    'Node1': [72600, 47369, 14728, 0, 10503],
    'Node2': [72600, 60213, 0, 12306, 0],
    'Node3': [72600, 53792, 18808, 0, 0]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Use Seaborn's Set2 color palette
palette = sns.color_palette("Set2", len(df['Category']))

# Loop through each category
for i, category in enumerate(df['Category']):
    plt.bar(df.columns[1:], df.iloc[i, 1:], bottom=df.iloc[:i, 1:].sum(axis=0),
            label=category, color=palette[i])

# Add labels and title
plt.xlabel('Nodes')
plt.ylabel('Percentage')
plt.title('Partition of Categories in Nodes')
plt.legend()

# Show plot
plt.show()
```

```python
# Calculate total for each category
df['Total'] = df[['Node1', 'Node2', 'Node3']].sum(axis=1)

# Calculate percentages
for node in ['Node1', 'Node2', 'Node3']:
    df[node] = df[node] / df['Total'] * 100

# Use Seaborn's Set2 color palette
palette = sns.color_palette("Set2", len(df['Category']))

# Define number of columns and rows for subplots
num_rows = 2
num_cols = 3

# Create subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 10))

# Flatten axes for easier iteration
axes = axes.flatten()

# Plot each category in its own subplot
for i, category in enumerate(df['Category']):
    row = i // num_cols
    col = i % num_cols
    ax = axes[i]
    
    ax.bar(df.columns[1:-1], df.iloc[i, 1:-1], color=palette[i])
    ax.set_title(category)
    ax.set_xlabel('Nodes')
    ax.set_ylabel('Percentage')

# Hide any extra subplots
for i in range(len(df['Category']), num_rows * num_cols):
    axes[i].axis('off')

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
