## First analysis 
The first part of the analysis has the main goal of observing how many types of attacks are represented and with which frequency. Moreover, there is a clear distinction between malign and benign traffic. 

By looking at the available files, we have the following types of attacks: 
* Backdoor attack: Exploit of an existing backdoor in the system, avoiding cybersecurity measures and gaining unauthorized access to the system (https://nordvpn.com/es/blog/backdoor-attack/). 
* DDoS HTTP Flood attack: Type of DDoS attack in which the targeted server is overwhelmed with HTTP requests (used to load contents from webpage or send contents from Internet). There are two types: HTTPS GET attacks (request to obtain a certain file or image from web) and POST attacks (sending forms with data - web would need to handle the form and store the data somewhere). (https://www.cloudflare.com/learning/ddos/http-flood-ddos-attack/) 
* DDoS ICMP Flood attack: Overwhelm with ICMP echo-request packets. (https://www.cloudflare.com/learning/ddos/ping-icmp-flood-ddos-attack/)

* DDoS UDP Flood attack:

### Attack traffic 

```python
# Obtain the attack traffic csv files to explore the data 
backdoor = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/Backdoor_attack.csv')
ddos_https = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/DDoS_HTTP_Flood_attack.csv')
ddos_icmp = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/DDoS_ICMP_Flood_attack.csv')
ddos_tcp = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/DDoS_TCP_SYN_Flood_attack.csv')
ddos_udp = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/DDoS_UDP_Flood_attack.csv')
mitm = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/MITM_attack.csv')
os_fing = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/OS_Fingerprinting_attack.csv')
psswd = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/Password_attack.csv')
prt_scan = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/Port_Scanning_attack.csv')
rnsmw = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/Ransomware_attack.csv')
sql = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/SQL_Injection_attack.csv')
upld = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/Uploading_attack.csv')
vul_scan = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/Vulnerability_scanner_attack.csv')
xss = pd.read_csv('C:/Users/UX430/Documents/thesis/datasets/EDGE-IIOTSET/Edge-IIoTset dataset/Attack traffic/XSS_attack.csv')
```

```python
# Observe if all csv files have the same structure (number of columns and types of columns)
a = pd.DataFrame([backdoor.columns, ddos_https.columns, ddos_icmp.columns, ddos_tcp.columns, ddos_udp.columns, mitm.columns , os_fing.columns , psswd.columns , prt_scan.columns , rnsmw.columns , sql.columns , upld.columns , vul_scan.columns , xss.columns])
all(a[column].nunique() == 1 for column in a.columns) #values in each col are the same == all csv files have same structure
```

```python
# Let's merge the csv files into one dataframe
attack_traffic = pd.concat([backdoor, ddos_https, ddos_icmp, ddos_tcp, ddos_udp, mitm, os_fing, psswd, prt_scan, rnsmw, sql, upld, vul_scan, xss], ignore_index=True)
```

```python
# Get possible values from each column
for column in attack_traffic.columns:
    print(column, attack_traffic[column].unique())
```

```python
# Get frequency of each attack using column Attack_type 
freq = attack_traffic['Attack_type'].value_counts()
freq
```

```python
# Plot frequency of each attack 
freq.plot(kind='bar', color='blue')
plt.xlabel('Attack Type')
plt.ylabel('Frequency')
plt.title('Frequency of Each Attack Type')
plt.show()
```
